```python
# Notebook: vendor_validation_realistic.ipynb
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# Simulate two vendors
num_patients = 10000
patient_ids = [f"P{100000+i}" for i in range(num_patients)]

def generate_vendor_data(vendor_name):
    num_records = 12000
    df = pd.DataFrame({
        'patient_id': np.random.choice(patient_ids, num_records),
        'procedure_code': np.random.choice(['PROC001','PROC002','PROC003','PROC004','PROC005'], num_records),
        'claim_amount': np.round(np.random.uniform(50,15000,num_records),2),
        'vendor': vendor_name
    })
    return df

vendor1_df = generate_vendor_data('Vendor1')
vendor2_df = generate_vendor_data('Vendor2')

# Compare vendors for mismatched claim amounts
merged = pd.merge(vendor1_df, vendor2_df, on=['patient_id','procedure_code'], suffixes=('_v1','_v2'))
mismatches = merged[np.abs(merged['claim_amount_v1'] - merged['claim_amount_v2']) > 1.0]  # >1$ difference

print("Number of mismatched claim amounts between vendors:", mismatches.shape[0])
```
